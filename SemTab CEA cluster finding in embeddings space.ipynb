```python
import os
import requests
import pandas as pd
from wikibase_api import Wikibase
wb = Wikibase()
S = requests.Session()


vectors = []
üversichtlich = dict()
allCandites = dict()
allVectors = dict()

#getting all candidates for the cells of each tables (5 per cell)
for tableName in tablesWithKnownEnts:
    table = pd.read_csv("D:/Masterarbeit/data/Tables/SemTab/SemTab2020_Table_GT_Target/Round1/tables/HOQXJ2CD.csv")
    print(table.head())
    for row in table["col0"]:
        searchResults = []
        wikidataLabelSearch = wb.entity.search(row, "en", limit = 5)
        for result in wikidataLabelSearch["search"]:
            searchResults.append(result["id"])
        allCandites[row] = searchResults
    break #only one table for now
    
#getting the embeddings for the candidates
for ent in allCandites:
    vectors = []
    responses = []
    for ids in allCandites.get(ent):
        vectorLookup = S.get("https://wembedder.toolforge.org/api/vector/"+str(ids))
        print(vectorLookup)
        if vectorLookup.ok:
            vectors.append(vectorLookup.json()['vector'])
            responses.append(ids)
    allVectors[ent] = vectors
    üversichtlich[ent] = responses
```
