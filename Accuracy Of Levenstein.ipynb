```python
import os
from Levenshtein import distance, ratio
import requests
import xml.etree.ElementTree as ET
import pandas as pd
from sklearn.metrics import accuracy_score


accuracies = dict()
errors = dict()
results = dict()
i = 0

for filename in os.listdir('D:/Masterarbeit/data/Goldstandard/complete_tables/entities'):
    if i == 50:
        break
    i = i+1
    lookUpEntetiesUri = []
    table = pd.read_csv('D:/Masterarbeit/data/Goldstandard/complete_tables/table/'+ filename)
    attri = pd.read_csv('D:/Masterarbeit/data/Goldstandard/complete_tables/attributes/'+ filename, header=None)
    entities = pd.read_csv('D:/Masterarbeit/data/Goldstandard/complete_tables/entities/'+ filename, header=None)
        
    keyAttribute = attri.loc[attri[2] == True,1].values[0]
    print("for table ", filename, "--->", keyAttribute, " is the key attribute")
    print(attri.head())
    try:
        for label in table[keyAttribute]:
            dbPediaLookupResp = ET.fromstring(requests.get('https://lookup.dbpedia.org/api/search?maxResults=5&query='+str(label)).text)
            try:
                lookUpEntetiesUri.append(str(max(dbPediaLookupResp.findall('Result'), key=lambda k: ratio(str(k[0].text), str(label)))[1].text))
            except ValueError:
                lookUpEntetiesUri.append(label + ": nicht gefunden")

        entitiesUri = [str(ent) for ent in entities[0]]
        tmp = entitiesUri.copy()
        entitiesUri.sort(key = lambda k: entities[2][tmp.index(k)])
        try:
            accuracies[filename] = accuracy_score(lookUpEntetiesUri, entitiesUri)
        except ValueError:
            print("couldn't compute accuracy for ",filename)
            continue
        results[filename] = [[k,j] for k,j in zip(lookUpEntetiesUri, entitiesUri)]
    except Exception:
        errors[filename] = Exception
        print("Ups, somthing went wrong")
        continue
```

```python
for acc in accuracies.values():
    print(acc)
    
print(len(accuracies.values()))
```
